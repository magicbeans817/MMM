# Problem 2 - Cobbweb model

My surname is "Suchánek". Therefore according to the table, a = 4; b = 1; c = 4; d = 0.4.
My personalized equation:


```python
t = sp.Symbol('t', integer = True)

demand = sp.Function('qd')
supply = sp.Function('qs')
price = sp.Function('p')

a = sp.symbols('a')
b = sp.symbols('b')
c = sp.symbols('c')
d = sp.symbols('d')

demand_equation = sp.Eq(demand(t), a - b * price(t))
supply_equation = sp.Eq(supply(t), c + d * price(t-1))

demand_equation_s = demand_equation.subs([(a,4),(b,1)])
supply_equation_s = supply_equation.subs([(c,4),(d,0.4)])

display(demand_equation_s, supply_equation_s)

print("In equilibrium holds:")

display(sp.Eq(demand(t), supply(t)))

print('Therefore:')

equation = sp.Eq(demand_equation_s.rhs, supply_equation_s.rhs)


display(equation)

print('This can be easily solved into this (hand written solution scanned):')

recursive = sp.Eq(price(t+1), -0.4*price(t))
solution = sp.Eq(price(t), sp.rsolve(equation, price(t)))
display(recursive, solution)
```

## Phase diagram

```python
# the gift
def path(fun, start, iter):
    x = np.zeros(2*iter+1) #numpy.zeros() - Return a new array of given shape and type, filled with zeros.
    y = np.zeros(2*iter+1)
    x[0] = start
    y[0] = 0
    i = 1
    while i < 2*iter:
        x[i] = x[i-1]
        y[i] = fun(x[i])
        x[i+1] = y[i]
        y[i+1] = y[i]
        i += 2
    u = np.zeros(len(x)-1)
    v = np.zeros(len(y)-1)
    for i in range(len(x)-1):
        u[i] = x[i+1] - x[i]
        v[i] = y[i+1] - y[i]
    return(x, y, u, v)
```

```python
# prepare a python function
recF = sp.lambdify(price(t), recursive.rhs, "numpy")
display(
    equation,
    inspect.findsource(recF)
    )
```

```python
# Python phase diagram
tval = np.linspace(-5, 5, 50)
x, y, u, v = path(recF, 4, 6)

plt.figure(figsize=(15,15))
plt.plot(tval, recF(tval))
plt.plot(tval, tval, color = 'orange')
plt.plot(tval, recF(tval), color="blue")
plt.vlines(0, -5, 5, color = 'black')
plt.hlines(0, -5, 5, color = 'black')
plt.title("Phase diagram of Cobbweb model")
plt.quiver(x[:-1],y[:-1], u, v,  angles='xy', scale_units='xy', scale=1, width = 0.002, color = 'red')
```

## Problem 4

My initials "JS" result in strange system, thus for the purpose of this problem, I use "AS" initials.

```python
t = sp.symbols('t', real=True)

x = sp.Function('x')(t)
y = sp.Function('y')(t)

a = sp.symbols('a')
b = sp.symbols('b')
c = sp.symbols('c')
d = sp.symbols('d')
e = sp.symbols('e')
f = sp.symbols('f')
eq1 = sp.Eq(x.diff(t), a*x + b*y + e)
eq2 = sp.Eq(y.diff(t), c*x + d*y + f)
eq1s = eq1.subs([(a,-3),(b,3),(e,-2)])
eq2s = eq2.subs([(c,-1),(d,2),(f,1)])
display(eq1s, eq2s)
```

```python
# These are the formulas for equilibrium lines
x0 = sp.solve(sp.Eq(eq1s.rhs), y)[0] #x_dot = 0
y0 = sp.solve(sp.Eq(eq2s.rhs), y)[0] #y_dot = 0

display(x0, y0)
```

```python
x_eq = sp.solve(sp.Eq(x0, y0), x)[0]
y_eq = y0.subs(x, x_eq)

display(x_eq, y_eq)
```

```python
eigens = sp.Matrix([
    sp.Poly(eq1s.rhs).coeffs()[:2],
    sp.Poly(eq2s.rhs).coeffs()[:2]
]).eigenvects()

display(eigens)

sol = sp.dsolve([eq1s, eq2s],[x, y])

display(sol)
```

```python
xsize = 25                                        # number of points on x axis
ysize = 25                                        # number of points on y axis
x_vals = np.linspace(-10, 10, xsize)              # prepare the x-values
y_vals = np.linspace(-30, 30, ysize)              # prepare the y-values

xeq = sp.lambdify(x, x0)
yeq = sp.lambdify(x, y0)
x0_vals = [xeq(i) for i in x_vals]                # equilibrium line x' = 0
y0_vals = [yeq(i) for i in x_vals]                # equilibrium line y' = 0
```

```python
def particular(eq, x_cond, y_cond, points = 150, tmin = -2, tmax = 2):
  
    xt0 = eq[0].rhs.subs(t,0)                                               # substitute t = 0 into both equations
    yt0 = eq[1].rhs.subs(t,0)
    
    conds = sp.solve([sp.Eq(xt0, x_cond), sp.Eq(yt0, y_cond)])              # find the parameter values
    c1 = sp.Symbol('C1')
    c2 = sp.Symbol('C2')
    
    sol_p = [i.subs([('C1', conds[c1]), ('C2', conds[c2])]) for i in eq]   # and substitute them back into equations

    t_vals = np.linspace(tmin, tmax, points)                                # define t interval
    x_p = np.zeros(points)
    y_p = np.zeros(points)

    xfun = sp.lambdify(t, sol_p[0].rhs)
    yfun = sp.lambdify(t, sol_p[1].rhs)
    
    for i in range(points):                                                  # and calculate x, y values
        x_p[i] = xfun(t_vals[i])
        y_p[i] = yfun(t_vals[i])
    
    return(x_p, y_p)
```

```python
xyp = [particular(sol, i[0], i[1]) for i in [(6,10),(-5,-3),(-8,-15),(2.5,2.3)]]
_ = [plt.plot(i[0], i[1], linewidth = 2, color = 'black') for i in xyp]

def vector_field(xfun, yfun, xvals, yvals):
    gridsize = len(xvals) * len(yvals)            # the grid [x,y]
    xa_vals = np.zeros(gridsize)                  # arrow position x
    ya_vals = np.zeros(gridsize)                  # arrow position y
    dxa = np.zeros(gridsize)                      # arrow direction x
    dya = np.zeros(gridsize)                      # arrow direction y
    
    p = 0
    for i in xvals:                               # for every x value
        for j in yvals:                           # for every x & y value
            xa_vals[p] = i                        # save the combination
            ya_vals[p] = j
            dxa[p] = xfun(i, j)                   # calculate both derivatives
            dya[p] = yfun(i, j)
            p += 1                                # increment the position
    
    return(xa_vals, ya_vals, dxa, dya)
```
